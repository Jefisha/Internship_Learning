```python
import chromadb
print(chromadb.__version__)
```

```python

# 1. Create a ChromaDB client (in-memory)
client = chromadb.Client()

# 2. Create a collection
collection = client.create_collection("test_collection")

# 3. Add some text data
collection.add(
    ids=["1", "2"],
    documents=["I love machine learning", "Python is amazing"]
)

# 4. Query similar text
result = collection.query(
    query_texts=["machine"],
    n_results=2
)

print(result)
```
