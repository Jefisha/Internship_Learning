```python
import chromadb
client=chromadb.Client()
```

```python
collections=client.create_collection(name="test_collection")
```

```python
collections.add(
    ids=["101", "102", "103"],
    documents=[
        "Machine learning is the study of patterns and predictions.",
        "Deep learning uses neural networks with multiple layers.",
        "Python is widely used in AI and data science."
    ],
    metadatas=[
        {"topic": "ML"},
        {"topic": "DL"},
        {"topic": "Python"}
    ]
)

print("Data inserted successfully!")
```

```python
# Semantic Search
# -----------------------------------------
print("\n--- Semantic Search (query: 'neural networks') ---")
results = collections.query(
    query_texts=["neural networks"],
    n_results=1
)
print(results)
```

```python
#  Metadata Filter Search
# -----------------------------------------
print("\n--- Metadata Filter: topic='Python' ---")
filtered = collections.query(
    query_texts=["programming"],
    where={"topic": "Python"},
    n_results=1
)
print(filtered)
```

```python
#  Update an entry
# -----------------------------------------
collections.update(
    ids=["103"],
    documents=["Python is the most popular language for AI applications"],
    metadatas=[{"topic": "Python"}]
)

print("\n--- After Update: Query 'AI language' ---")
update_result = collections.query(
    query_texts=["AI language"],
    n_results=2
)
print(update_result)
```

```python
#  Delete an entry
# -----------------------------------------
collections.delete(ids=["102"])
print("\nDeleted document with id 102.")

print("\n--- Final data check ---")
final_check = collections.query(
    query_texts=["anything"],
    n_results=5
)
print(final_check)

```
